```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy import integrate 
```

```python
def f(y,x):
    return np.sin(x)
```

```python
def ex(x):
    return 2-np.cos(x)
```

```python
f0=1.0
x=np.linspace(0,10,100)
```

```python
sol=integrate.odeint(f,f0,x)
```

```python
fig, axes=plt.subplots()
axes.plot(x,ex(x))
plt.show()
```

```python
fig, axes=plt.subplots()
axes.plot(x,sol, '--')
plt.show()
```
