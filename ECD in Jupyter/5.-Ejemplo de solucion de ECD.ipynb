```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Funcion que retorna dy/dt
def model(y,t):
    k = 0.3
    dydt = -k * y
    return dydt

# Condicion Inicial 
y0 = 5

# Puntos en relacion al tiempo 
t = np.linspace(0,20)

# Resolucion de EDO 
y = odeint(model,y0,t)

# Graficacion 
plt.plot(t,y)
plt.xlabel('tiempo')
plt.ylabel('y(t)')
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Funcion que retorna dy/dt
def model(y,t,k):
    dydt = -k * y
    return dydt

# Condicion inicial 
y0 = 5

# Puntos en el tiempo 
t = np.linspace(0,20)

# Resolucion de ECOs 
k = 0.1
y1 = odeint(model,y0,t,args=(k,))
k = 0.2
y2 = odeint(model,y0,t,args=(k,))
k = 0.5
y3 = odeint(model,y0,t,args=(k,))

# Graficacion 
plt.plot(t,y1,'r-',linewidth=2,label='k=0.1')
plt.plot(t,y2,'b--',linewidth=2,label='k=0.2')
plt.plot(t,y3,'g:',linewidth=2,label='k=0.5')
plt.xlabel('tiempo')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```
